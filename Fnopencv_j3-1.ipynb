```python
def imshow(image):
    cv2.imshow('ax1',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
```

```python
ret,img_tr=cv2.threshold(img_gray,100,255,0)
imshow(img_tr)
print(ret)
```
